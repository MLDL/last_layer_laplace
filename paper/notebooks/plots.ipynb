```python
%matplotlib inline
import numpy as np
import matplotlib
import seaborn as sns
import tikzplotlib

matplotlib.rcParams['figure.figsize'] = (5,5)
matplotlib.rcParams['font.size'] = 20
matplotlib.rcParams['font.family'] = "serif"
matplotlib.rcParams['font.serif'] = 'Times'
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['lines.linewidth'] = 3
plt = matplotlib.pyplot

sns.set('talk')
sns.set_style('white')
```

```python
x = np.linspace(-10, 10, 1000)
```

```python
np.random.seed(45)

for _ in range(5):
    plt.plot(x, np.random.randn()*x + np.random.randn()*3, c='k', alpha=0.25)
    
point = np.array([0.1, 0.1])
plt.plot(*point, c='k', marker='o', zorder=9999)
plt.text(0.75, -0.5, r'$\mathbf{x}$')

a = 70*point-point
plt.arrow(*point, *a, head_width=0.5, head_length=0.5, fc='k', ec='k')

plt.plot(4.6, 4.6, marker='o')
plt.text(4.6, 3, r'$\alpha \mathbf{x}$')
plt.text(*70*point+0.5, r'$\delta \mathbf{x}$')
    
plt.xlim(-10, 10);
plt.ylim(-10, 10);
plt.xticks([]);
plt.yticks([]);

tikzplotlib.save('figs/nonasymp_situation.tex')
```
